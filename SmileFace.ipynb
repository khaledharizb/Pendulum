```julia
using Plots


function SmileFace(R, u0)
step=π / 1000;    
t=0:step:2*π; n=length(t) ; a = 0.1;
    
y= hcat(u0[1] .+ R*cos.(t),u0[2] .+R*sin.(t)) # big circle of radius
y1=hcat(u0[1] -a  .+ R/8*cos.(t),u0[2] + 0.15 .+ R/5*sin.(t))  # left small circle
y2=hcat(u0[1] +a  .+ R/8*cos.(t),u0[2] + 0.15 .+ R/5*sin.(t))  # right small circle


a =0.3
    
t = range(u0[1]+a, u0[1]-a, 400)

y3= hcat(t, 3.5*R*(t .-u0[1]).^2 .+ u0[2] .- 0.2) #  parabola

    
Y = [y; NaN NaN; y1; NaN NaN; y2; NaN NaN; y3]
    p=plot(aspect_ratio=1,framestyle=:none,legend=:false)
  plot!(p,Y[1:n,1],Y[1:n,2],lc=:black,fillcolor=:yellow,lw = 2,fill=true)
  plot!(p,Y[n+1:3*n,1],Y[n+1:3*n,2],lc=:black,fillcolor=:black,lw = 2,fill=true)  
  plot!(p,Y[3*n+2:end,1],Y[3*n+2:end,2],lc=:black,lw = 3,fill=false)    
    return p
end

R=0.4; u0=[1.2,0];

SmileFace(R, u0)
```
